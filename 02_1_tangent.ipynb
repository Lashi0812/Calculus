```python
import sympy as sp
import spb

x,y,m = sp.symbols("x y m")
```

# slope of Tangent

Slope of tangent is limit of the slope of secant.

$$\lim_{Q \to P} m_{PQ} = m$$

```python
fx = x**2
fx
```

```python
slope_of_secant = (fx - fx.subs(x,1))/(x-1)
slope_of_secant
```

```python
m_pq = slope_of_secant.subs(x,2)
m_pq
```

```python
secant_line_eq = m_pq*(x-2) + fx.subs(x,2)
secant_line_eq
```

```python
slope_of_tangent = sp.limit(slope_of_secant,x,1)
slope_of_tangent
```

```python
tangent_line_eq = slope_of_tangent*(x-1) + fx.subs(x,1)
tangent_line_eq
```

```python
p1 = spb.plot(fx,(x,-3,3),show=False)
p2 = spb.plot_list(([1,2],[fx.subs(x,1),fx.subs(x,2)]),show=False,is_filled=True,scatter=True)
secant_line_plot = spb.plot(secant_line_eq,(x,0,3),show=False,label="secant_line") 
tangent_line_plot = spb.plot(tangent_line_eq,(x,0,3),show=False,label="tangent_line") 
(p1 + p2 + secant_line_plot + tangent_line_plot).show()
```
