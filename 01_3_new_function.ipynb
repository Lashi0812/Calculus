```python
import sympy as sp
import spb
from matplotlib.gridspec import GridSpec
```

# Translation

```python
x,c = sp.symbols("x c")
imodule = "panel"
options = dict(
    imodule=imodule, show=False, use_latex=False, params={
        c: (0, 0,5),
    },legend=True)


p1 = spb.plot(x**2+c, (x, -2, 2), **options)
p2 = spb.plot(x**2-c, (x, -2, 2), **options)
p3 = spb.plot((x+c)**2, spb.prange(x, -c-2, c+2), **options)
p4 = spb.plot((x-c)**2, spb.prange(x, -c-2, c+2), **options)

gs = GridSpec(2, 2)
mapping = {
    gs[0,0] : p1,   
    gs[0,1] : p2,
    gs[1,0] : p3,
    gs[1,1] : p4,
}
spb.plotgrid(gs=mapping, imodule=imodule, use_latex=False)
```

```python
p1 = sp.plot(x**2,(x,-2,2),show=False)
p2 = sp.plot(x**2 - 1,(x,-2,2),show=False)
p3 = sp.plot(sp.Abs(x**2 - 1),(x,-2,2),show=False)

sp.plotting.PlotGrid(1,3,p1,p2,p3)
```

$f(x) = \sqrt{25 - x^{2}} , g(x) = \sqrt{x + 1}$

```python
x = sp.symbols("x")
a = 25 - x**2
fx = sp.sqrt(a)
fx
```

```python
b = x+1
gx = sp.sqrt(b)
gx
```

```python
f_domain = sp.solve_univariate_inequality(sp.Ge(a, 0), x, relational=False)
f_domain
```

```python
g_domain = sp.solve_univariate_inequality(sp.Ge(b,0),x,relational=False)
g_domain
```

```python
f_domain.intersection(g_domain)
```

```python
hx = fx + gx
hx
```

```python
sp.calculus.util.continuous_domain(hx,x,sp.S.Reals)
```
