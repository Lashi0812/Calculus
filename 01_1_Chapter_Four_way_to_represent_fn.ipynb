# Determine whether the equation or table defines y as a function of x

## $3x - 5y = 7$

```python
# Generate x values
x = np.linspace(-5, 5, 400)

# Calculate corresponding y values using the given equation
y = (7-3*x)/5

# Plot the curve
plt.plot(x, y, label='$3x - 5y = 7$')

# Label the axes and add a legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show the plot
plt.grid(True)
plt.axis('equal')
plt.show()
```

## $f(x)=\frac{x+4}{x^{2}-9}$

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x = sp.Symbol("x")
fx_denominator = x**2 -9
fx = (x + 4) / fx_denominator
```

```python
domain  = sp.S.Reals -set(sp.real_roots(fx_denominator,x))
domain
```

```python
sp.plot(fx,(x,-10,10),ylim=(-2,2),xlim=(-10,10))
```

## $h(x)=\frac{1}{\sqrt[4]{x^{2}-5x}}$

```python
a = x**2 - 5*x
b = sp.root(a,4)
fx = 1/b
fx
```

* $x^{2}-5x \neq 0$  --> this can be done by solving for roots  
* $x^{2}-5x > 0$     --> this can be done by solving for inequality

```python
sp.real_roots(a,x)
```

```python
sp.solve_univariate_inequality(a > 0,x,relational=False)
```

```python
sp.calculus.util.continuous_domain(fx,x,sp.S.Reals)
```

```python
# we will runtime waring since we are including the x [0,5]
sp.plot(fx,(x,-10,10))
```

```python
sp.plot((fx,(x,-10,0)),(fx,(x,5,10)),ylim=(0,5))
```

## $f(u)=\frac{u+1}{1+\frac{1}{u+1}}$

```python
a = x+1
b = 1/a
c = 1+b
fx = a/c
fx
```

$x+1 \neq 0$ and ${1+\frac{1}{u+1}} \neq 0$

```python
sp.simplify(c)
```

```python
sp.real_roots(a,x)
```

```python
sp.solve(c,x)
```

```python
sp.S.Reals - set(sp.solve(a,x)) - set(sp.solve(c,x))
```

```python
sp.calculus.util.continuous_domain(fx,x,sp.S.Reals)
```

```python
sp.plot(fx,(x,-10,10))
```

```python
sp.plot(
    (fx,(x,-10,-2)),
    (fx,(x,-2,-1)),
    (fx,(x,-1,10)),
    ylim = (-100,100)
)
```
