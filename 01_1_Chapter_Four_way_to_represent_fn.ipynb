# Determine whether the equation or table defines y as a function of x

## $3x - 5y = 7$

```python
# Generate x values
x = np.linspace(-5, 5, 400)

# Calculate corresponding y values using the given equation
y = (7-3*x)/5

# Plot the curve
plt.plot(x, y, label='$3x - 5y = 7$')

# Label the axes and add a legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show the plot
plt.grid(True)
plt.axis('equal')
plt.show()
```

## $(3x^2 - 2y = 5)$

```python
x = np.linspace(-5,5,400)
y = (5 - 3*x**2)/2
plt.plot(x,y)
plt.grid(True)
plt.title("$(3x^2 - 2y = 5)$")
plt.show()
```

## $(x^2 + (y-3)^2 = 5)$

this is not a function as it doesn't pass vertical line test.

```python
y1 = 3 + np.sqrt(5 - x**2)
y2 = 3 - np.sqrt(5 - x**2)

x = np.linspace(-2.2,2.2,400)
plt.plot(x,y1)
plt.plot(x,y2)
plt.title("$x^2 + (y-3)^2 = 5$")
plt.show()
```

# $f(x) = 3x^2 - x +2$

```python
import sympy

x = sympy.Symbol("x")
fx = 3*x**2 - x + 2
fx
```

```python
fx.subs(x,2)
```

```python
fx.subs(x,-2)
```

```python
fx.subs(x,"a")
```

```python
fx.subs(x,"-a")
```

```python
sympy.expand(fx.subs(x,"a+1"))
```

```python
2*fx.subs(x,"a")
```

```python
fx.subs(x,"2*a")
```

```python
fx.subs(x,"a**2")
```

```python
sympy.expand(fx.subs(x,"a")**2)
```

```python
sympy.expand(fx.subs(x,"a+h"))
```
