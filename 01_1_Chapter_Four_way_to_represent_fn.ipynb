# Determine whether the equation or table defines y as a function of x

## $3x - 5y = 7$

```python
# Generate x values
x = np.linspace(-5, 5, 400)

# Calculate corresponding y values using the given equation
y = (7-3*x)/5

# Plot the curve
plt.plot(x, y, label='$3x - 5y = 7$')

# Label the axes and add a legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show the plot
plt.grid(True)
plt.axis('equal')
plt.show()
```

## $(3x^2 - 2y = 5)$

```python
x = np.linspace(-5,5,400)
y = (5 - 3*x**2)/2
plt.plot(x,y)
plt.grid(True)
plt.title("$(3x^2 - 2y = 5)$")
plt.show()
```

## $f(x)=\frac{x+4}{x^{2}-9}$

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x = sp.Symbol("x")
fx_denominator = x**2 -9
fx = (x + 4) / fx_denominator
```

```python
domain  = sp.S.Reals -set(sp.real_roots(fx_denominator,x))
domain
```

## $h(x)=\frac{1}{\sqrt[4]{x^{2}-5x}}$

```python
a = x**2 - 5*x
b = sp.root(a,4)
fx = 1/b
fx
```

* $x^{2}-5x \neq 0$  --> this can be done by solving for roots  
* $x^{2}-5x > 0$     --> this can be done by solving for inequality

```python
sp.real_roots(a,x)
```

```python
sp.solve_univariate_inequality(a > 0,x,relational=False)
```

```python
sp.calculus.util.continuous_domain(fx,x,sp.S.Reals)
```

```python
# we will runtime waring since we are including the x [0,5]
sp.plot(fx,(x,-10,10))
```

```python
sp.plot((fx,(x,-10,0)),(fx,(x,5,10)),ylim=(0,5))
```

```python
a = x+1
b = 1/a
c = 1+b
fx = a/c
fx
```

## 49.
$49. f(x)=\begin{cases} x^2+2 & \text{if } x<0 \ \ x & \text{if } x\ge0\end{cases}$

```python
a = x**2 + 2
b = x
fx = sp.Piecewise((a,x<0),(b,x>=0))
fx
```

```python
print(f"f(-3) = {fx.subs(x,-3)} ")
```

```python
print(f"f(0) = {fx.subs(x,0)} ")
```

```python
print(f"f(2) = {fx.subs(x,2)} ")
```

```python
sp.plot(
    (a,(x,-2,0)),
    (b,(x,0,2)),
    legend=True
)
```
