```python
import sympy as sp
import spb

x,a = sp.symbols("x a")
```

# Example 1

```python
fx = (sp.sqrt(x**2 + 9) - 3) / x**2
fx
```

```python
sp.limit(fx,x,0)
```

```python
imodule = "panel"
options = dict(
    imodule=imodule, use_latex=False, params={
        a: (5, 0.0000001,1),
    },legend=True)
spb.plot(fx, spb.prange(x, -a, a), **options)
```

# one sided limits

```python
fx = sp.Piecewise((0,sp.Lt(x,0)),(1,sp.Ge(x,0)))
fx
```

```python
spb.plot_piecewise(fx,(x,-2,2))
```
